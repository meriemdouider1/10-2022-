```python
def transform(X, fs) :
  X_ = X.copy()
  for i in range(len(fs)) :
    if fs[i] == False :
      X_.drop([clmns[i]], axis=1, inplace=True)
  return X_
```

```python
def Retransform (lst):
  Fs=[]      
  for i in range(225):
    if clmns[i] in lst :
       Fs.append(True)
    else :
       Fs.append(False)
  return Fs
```

```python
cv = KFold(n_splits=5, random_state=1, shuffle=True)
explored_models_mcc = []
scores_mcc = []
models_mm_mcc=[]
var_mcc=[]
def explor_mcc(fs, mcc, algo) :
  if (sum(fs) > 225) :
    return
  explored_models_mcc.append(fs)
  for i in range(225) :
    neighbor = fs.copy()
    neighbor[i] = not fs[i]
    if (neighbor not in explored_models_mcc) :
      score_mcc = cross_val_score(algo, transform(X, neighbor) , y, scoring=my_scorer1, cv=cv, n_jobs=-1)
      mean_score_mcc = np.mean(score_mcc)
      explored_models_mcc.append(neighbor)
      if (mean_score_mcc  > mcc) :
        models_mm_mcc.append(neighbor)
        print("mcc : %f" % mean_score_mcc)
        print(len(transform(X,neighbor).columns))
        print(transform(X,neighbor).columns)
        print("-----------")
        scores_mcc.append(mean_score_mcc)
        var_mcc.append(transform(X,neighbor))
        explor_mcc(neighbor, mean_score_mcc, algo)
```
