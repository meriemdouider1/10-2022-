```python
def specificite(y_true, y_pred):
  specificite= confusion_matrix(y_true, y_pred)[0,0]/(confusion_matrix(y_true,y_pred)[0,0] + confusion_matrix(y_true, y_pred)[0,1])
  return specificite
my_scorer = make_scorer(specificite, greater_is_better=True)
def MCC(y_true, y_pred):
  MCC= matthews_corrcoef(y_true, y_pred)
  return MCC
my_scorer1 = make_scorer(MCC, greater_is_better=True)
def présicion_positive(y_val, preds):
  cf_matrix = confusion_matrix(y_val, preds)
  pre_pos =  (cf_matrix[1,1])/(cf_matrix[1,1]+cf_matrix[0,1])
  return pre_pos
my_scorer2 = make_scorer(présicion_positive, greater_is_better=True)
def présicion_négative(y_val, preds):
  cf_matrix = confusion_matrix(y_val, preds)
  pre_pos =  (cf_matrix[0,0])/(cf_matrix[0,0]+cf_matrix[1,0])
  return pre_pos
my_scorer3 = make_scorer(présicion_négative, greater_is_better=True)
def VN(y_true, y_pred):
    value = confusion_matrix(y_true, y_pred)[0,0]
    return value
my_scorer4 = make_scorer(VN, greater_is_better=True)
def VP(y_true, y_pred):
    value = confusion_matrix(y_true, y_pred)[1,1]
    return value
my_scorer5 = make_scorer(VP, greater_is_better=True)
def FP(y_true, y_pred):
    value = confusion_matrix(y_true, y_pred)[0,1]
    return value
my_scorer6 = make_scorer(FP, greater_is_better=True)
def FN(y_true, y_pred):
    value = confusion_matrix(y_true, y_pred)[1,0]
    return value
my_scorer7 = make_scorer(FN, greater_is_better=True)


```

```python
def get_models():
 models = list()
 models.append(('GB', GradientBoostingClassifier(n_estimators=225, learning_rate=0.1, max_features=2, max_depth=2, random_state=0,min_samples_leaf=2,subsample=1 )))
 models.append(('XG', XGBClassifier(scale_pos_weight=1.5,learning_rate=0.1,max_depth=5,n_estimators=100)))
 models.append(('RF', RandomForestClassifier(n_estimators=150, criterion='gini', max_depth=5, random_state =11)))
 models.append(('GaussianNB' , GaussianNB()))
 models.append(("KNeighborsClassifier", KNeighborsClassifier(n_neighbors=4)))
 models.append(('SVC',make_pipeline(StandardScaler(), NuSVC())))
 return models

def evaluate_models_train(models, X_train, y_train):
 # fit and evaluate the models
 scores = list()
 for name, model in models:
   # fit the model
   model.fit(X_train, y_train)
   # evaluate the model
   yhat = model.predict(X_train)
   acc = accuracy_score(y_train, yhat)
   # store the performance
   scores.append(acc)
 # report model performance
 return scores

def evaluate_models_test(models, X_train, y_train,X_test,y_test):
 # fit and evaluate the models
 scores = list()
 for name, model in models:
   # fit the model
   model.fit(X_train, y_train)
   # evaluate the model
   yhat = model.predict(X_test)
   acc = accuracy_score(y_test, yhat)
   # store the performance
   scores.append(acc)
 # report model performance
 return scores

def predict_test(models, X_train, y_train,X_test,y_test):
  d=dict()
  for name, model in models:
   model.fit(X_train, y_train)
   # evaluate the model
   yhat = model.predict(X_test)
   d[name]= yhat
  return d

def weight_models (models,X_train, y_train, nb):
  score_acc = evaluate_models_train(models, X_train, y_train)
  score_acc_sum=sum(score_acc)
  weight=dict()
  for i in range(nb):
    a = score_acc[i]/score_acc_sum
    weight[models[i][0]] = a
  return weight

def voting (models, X_train, y_train,X_test,y_test, nb):
  weight = weight_models (models,X_train, y_train, nb)
  predict_t = predict_test(models, X_train, y_train,X_test,y_test)
  predict_ensemble=[]
  for l in range(len(y_test)):
    classe_0 = 0
    classe_1 = 0 
    for i in range(nb):
      cl=models[i][0]
      if predict_t[cl][l] == 1 :
        classe_1 = classe_1 + weight[cl]
      else :
        classe_0 = classe_0 + weight[cl]
    if(classe_1 > classe_0):
      predict_ensemble.append(1)
    else :
      predict_ensemble.append(0)
  return predict_ensemble


models = get_models()
def Ensmeble_voting(models,nb):
  cv = KFold(n_splits=5, random_state=1, shuffle=True)
  lst=[]
  acc_models_GB_test=[]
  acc_models_XG_test=[]
  acc_models_RF_test=[]
  acc_models_GN_test=[]
  acc_models_KNN_test=[]
  acc_models_SVC_test=[]
  for train_index, test_index in cv.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    evaluate_models_TRAIN = evaluate_models_train(models, X_train, y_train)
    evaluate_models_TEST =  evaluate_models_test(models, X_train, y_train,X_test,y_test)
    acc_models_GB_test.append(evaluate_models_TEST[0])
    acc_models_XG_test.append(evaluate_models_TEST[1])
    acc_models_RF_test.append(evaluate_models_TEST[2])
    acc_models_GN_test.append(evaluate_models_TEST[3])
    acc_models_KNN_test.append(evaluate_models_TEST[4])
    acc_models_SVC_test.append(evaluate_models_TEST[5])
    weight= weight_models (models,X_train, y_train, nb)
    print("-------------------------------------")
    print("ACCURACY_TRAIN",evaluate_models_TRAIN)
    print("ACCURACY_TEST",evaluate_models_TEST)
    print("-------------------------------------")
    print("weight", weight)
    print("-------------------------------------")
    Predict_Test_base_models = predict_test(models, X_train, y_train,X_test,y_test)
    print("Predict_Test_base_models",Predict_Test_base_models)
    print("-------------------------------------")
    predict_ensemble_test = voting (models, X_train, y_train,X_test,y_test, nb)
    print("predict_ensemble_test", predict_ensemble_test)
    print("-------------------------------------")
    print(len(predict_ensemble_test))
    print(len(y_test))
    print("-------------------------------------")
    acc = accuracy_score(y_test, predict_ensemble_test)
    lst.append(acc)
  return lst,acc_models_GB_test,acc_models_XG_test,acc_models_RF_test,acc_models_GN_test,acc_models_KNN_test,acc_models_SVC_test
    

acc_ensemble,acc_models_GB_test,acc_models_XG_test,acc_models_RF_test,acc_models_GN_test,acc_models_KNN_test,acc_models_SVC_test= Ensmeble_voting(models,6)
```
